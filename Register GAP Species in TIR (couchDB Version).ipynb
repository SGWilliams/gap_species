```python
import requests
import pysb
import time
from IPython.display import display
from bis2 import couchconnect
from bis import gap
import couchdb
```

```python
couch = couchdb.Server()

dbname = "tir-gapspecies"
if dbname in couch:
    db = couch[dbname]
else:
    db = couch.create(dbname)
```

```python
sb = pysb.SbSession()
username = input("Username: ")
sb.loginc(str(username))
```

```python
_habitatMapCollectionID = "527d0a83e4b0850ea0518326"
_gapSpecies = sb.get_child_ids(_habitatMapCollectionID)
```

```python
for item in _gapSpecies:
    if item not in db:
        print (item, "NEW")
        gapSpecies = gap.gapToTIR_flat(sb.get_item(item, {'fields':'identifiers,tags'}))
        db[item] = gapSpecies
    else:
        print (item, "EXISTING")
```
