I'm experimenting with a whole new way of running the Taxa Information Registry using a combination of NoSQL technologies. I've currently got foundational processes generating documents in a cloud-based MongoDB solution from MLab (sandbox). We'll move that over to our own beta and production instances sometime soon. I'm also experimenting with an cloud-based ElasticSearch sandbox from Bonsai for building up purpose-built indexes with TIR information for GAP species.

This notebook runs through a summary of what goes into the TIR for GAP species. I dumped an alternative output of the GAP species records as documents in a gapspecies collection for ease of use and because ScienceBase kept giving me problems. These are based on what we consider foundational records for the GAP species in the still internal ScienceBase collection for the habitat maps. The documents in the MongoDB collection contain the original ScienceBase document but also flatten the structure a little bit into top-level keys for ease of use (e.g., key identifiers are each separate attributes in the documents).

I'm working up a new process that will first set up processing in the various TIR caches (ITIS, WoRMS, TESS, and NatureServe) and then run those processes to retrieve and cache relevant documents. Those documents can then be used to assemble data for use in various places. I'm going to see what all we can do with ElasticSearch indexes, but we'll also use this process to feed specific attributes back into relevant ScienceBase items for GAP species habitat models and ranges so that they can work in search faceting through that system.

```python
import requests
from IPython.display import display
from bis2 import mlab
```

Here, we set up the connection to the MLab-based database (bis) and make connections for the various collections we are working with. The MLab connection requires a security key available from the secure bis2 package.

```python
bisDB = mlab.getDB("bis")
gapspecies = bisDB["gapspecies"]
itiscache = bisDB["itiscache"]
wormscache = bisDB["wormscache"]
tesscache = bisDB["tesscache"]
natureservecache = bisDB["natureservecache"]
iucncache = bisDB["iucncache"]
bisoncache = bisDB["bisoncache"]
```

Here, we loop through a limited number of the GAP species records, retrieve related documents from the various caches, and display all of that available information in its native JSON format. The point here is to simply see a picture of what's going on with this overall aspect of the Biogeographic Information System.

```python
for gapDoc in gapspecies.find().limit(20):
    print ("----------", gapDoc["GAP_SpeciesCode"], "/", gapDoc["scientificname"], "/", gapDoc["commonname"], "----------")
    
    display (gapDoc)

    itisDoc = itiscache.find_one({"originID":gapDoc["_id"]})
    wormsDoc = wormscache.find_one({"originID":gapDoc["_id"]})
    tessDoc = tesscache.find_one({"originID":gapDoc["_id"]})
    natureServeDoc = natureservecache.find_one({"originID":gapDoc["_id"]})
    iucnDoc = iucncache.find_one({"originID":gapDoc["_id"]})
    bisonDoc = bisoncache.find_one({"originID":gapDoc["_id"]})

    if itisDoc is not None:
        print ("---------- Cached Integrated Taxonomic Information System (ITIS) Document ----------")
        display (itisDoc)

    if wormsDoc is not None:
        print ("---------- Cached World Register of Marine Species (WoRMS) Document ----------")
        display (wormsDoc)

    if tessDoc is not None:
        print ("---------- Cached FWS Threatened & Endangered Species System (TESS) Document ----------")
        display (tessDoc)

    if natureServeDoc is not None:
        print ("---------- Cached NatureServe Document ----------")
        display (natureServeDoc)

    if iucnDoc is not None:
        print ("---------- Cached IUCN Species Document ----------")
        display (iucnDoc)

    if bisonDoc is not None:
        print ("---------- Cached BISON Document ----------")
        display (bisonDoc)

    print ("----------------------------------------------------------")
```
